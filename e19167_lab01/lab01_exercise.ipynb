```python
import numpy as np
import matplotlib.pyplot as plt

def imcomplement(I):
    return np.max(I) - I

def grayscale(array):
    # Calculate the average of the red, green, and blue channels
    return np.dot(array[...,:3], [0.33, 0.33, 0.33])

# Load the image
path = '/content/image2.jpg'
image_array = plt.imread(path)

# Invert the image


# Convert the image to grayscale
grayscale_image = grayscale(image_array)
inverted_image = 255 - grayscale_image
# Display the original, inverted, and grayscale images
fig, ax = plt.subplots(1, 3, figsize=(15, 5))

ax[0].imshow(image_array )
ax[0].set_title('Original Image')

ax[1].imshow( inverted_image, cmap='gray' )
ax[1].set_title('Inverted Image')

ax[2].imshow(grayscale_image, cmap='gray')
ax[2].set_title('Grayscale Image')

plt.show()
```

```python

flippedud = np.flipud(grayscale_image)
fig, ax = plt.subplots(1, 2, figsize=(10, 5))
ax[0].imshow( grayscale_image, cmap='gray')
ax[0].set_title('Original Image')

ax[1].imshow( flippedud, cmap='gray' )
ax[1].set_title('Rotated Image')
plt.show()
```

```python
flippedlr = np.fliplr(grayscale_image)
fig, ax = plt.subplots(1, 2, figsize=(10, 5))
ax[0].imshow( grayscale_image, cmap='gray')
ax[0].set_title('Original Image')

ax[1].imshow( flippedlr, cmap='gray' )
ax[1].set_title('Rotated Image')
plt.show()
```

```python
import numpy as np
import cv2
from google.colab.patches import cv2_imshow

# Read the image
path = '/content/image2.jpg'
image = cv2.imread(path)

# Translation
def translate(image, tx, ty):
    M = np.float32([[1, 0, tx], [0, 1, ty]])
    translated_image = cv2.warpAffine(image, M, (image.shape[1], image.shape[0]))
    return translated_image

# Similarity
def similarity(image, angle, scale):
    center = (image.shape[1] / 2, image.shape[0] / 2)
    M = cv2.getRotationMatrix2D(center, angle, scale)
    rotated_image = cv2.warpAffine(image, M, (image.shape[1], image.shape[0]))
    return rotated_image

# Projective (Perspective)
def projective(image, pts1, pts2):
    M = cv2.getPerspectiveTransform(pts1, pts2)
    transformed_image = cv2.warpPerspective(image, M, (image.shape[1], image.shape[0]))
    return transformed_image

# Euclidean
def euclidean(image, angle):
    center = (image.shape[1] / 2, image.shape[0] / 2)
    M = cv2.getRotationMatrix2D(center, angle, 1)
    rotated_image = cv2.warpAffine(image, M, (image.shape[1], image.shape[0]))
    return rotated_image

# Test the transformations
translated_image = translate(image, 100, 100)
rotated_image = similarity(image, 45, 1)
projective_image = projective(image, np.float32([[50, 50], [200, 50], [50, 200], [200, 200]]), np.float32([[10, 100], [200, 50], [100, 250], [200, 200]]))
euclidean_image = euclidean(image, 30)

# Display the images
cv2_imshow(image)




cv2.waitKey(0)
cv2.destroyAllWindows()
```
